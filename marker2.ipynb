```python
spot = input("가고 싶은곳을 입력하세요: ")
scale=input("거리를 입력하세요(m): ")
mon = input("예산을 입력하세요(원)")
scale1=int(scale)*1.25
sp=spot.split(",")
res=[]
lat,lon=0,0
lati,long=[],[]
for i in range(0,len(sp)):
    res.append(play[play['상호명']==sp[i]])
    tem_la=(res[i]['위도'])
    lat+=tem_la.values
    lati.append(tem_la.values)
    tem_lo=(res[i]['경도'])
    lon+=tem_lo.values
    long.append(tem_lo.values)
   
    folium.Circle([tem_la.values,tem_lo.values], radius = 300,color='green').add_to(m)
#m
lat/=len(sp)
lon/=len(sp)
folium.Circle([lat,lon],radius=scale1, color='yellow',fill_color='#fffggg').add_to(m)
```
